```python
matrix_size = (50, 50)


state_matrix = np.loadtxt('SwanLake_InitialCondition.csv', delimiter=',')

data_matrix = Matrix.run_simple_parameter_sweep(1, 0.0, 1.0, 101, state_matrix, 0, False, 17, 15, False, True)
```

```python

x_values = np.linspace(0, 100, 101)

y_values_burn_region = np.zeros_like(x_values)

k = 0
for i in data_matrix:
    for j in data_matrix[i]:
        y_values_burn_region[k] += j.burn_region
    y_values_burn_region[k] /= len(data_matrix[i])
    k += 1

k = 0
y_values_burn_time = np.zeros_like(x_values)
for i in data_matrix:
    for j in data_matrix[i]:
        y_values_burn_time[k] += j.burn_time
    y_values_burn_time[k] /= len(data_matrix[i])
    k += 1

k = 0
y_values_spread_rate = np.zeros_like(x_values)
for i in data_matrix:
    for j in data_matrix[i]:
        y_values_spread_rate[k] += j.spread_rate
    y_values_spread_rate[k] /= len(data_matrix[i])
    k += 1

print(y_values_burn_time)
print(y_values_burn_region)
print(y_values_spread_rate)

state_matrix
```

```python
for i in data_matrix:
    print(f"\nP_burn = {i}\n")
    for j in data_matrix[i]:
        print(j)
```

```python
fig = plt.figure(figsize=(17, 5))

plt.subplot(1, 3, 1)
plt.plot(x_values, y_values_burn_region)
plt.title('Average Burn Region (% of Total Area) vs. Probability of Burning (%)')
plt.xlabel('Probability of Burning (%)')
plt.ylabel('Average Burn Region (% of Total Area)')

plt.subplot(1, 3, 2)
plt.plot(x_values, y_values_burn_time)
plt.title('Burn Duration vs. Probability of Burning (%)')
plt.xlabel('Probability of Burning (%)')
plt.ylabel('Burn Duration')

plt.subplot(1, 3, 3)
plt.plot(x_values, y_values_spread_rate)
plt.title('Average Spread Rate vs. Probability of Burning (%)')
plt.xlabel('Probability of Burning (%)')
plt.ylabel('Average Spread Rate')

plt.tight_layout()
plt.show()
```
